```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
df.to_csv("restaurant_sales.csv", index=False)
```

```python
from google.colab import files
files.download("restaurant_sales.csv")
```

```python
df.describe()
```

```python
# Total revenue
total_revenue = df["total_bill"].sum()
print("Total Revenue:", total_revenue)
```

```python
# Average bill
avg_bill = df["total_bill"].mean()
print("Average Bill:", avg_bill)
```

```python
# Average tip
avg_tip = df["tip"].mean()
print("Average Tip:", avg_tip)
```

```python
# Revenue by day
revenue_by_day = df.groupby("day")["total_bill"].sum()
print(revenue_by_day)
```

```python
plt.figure(figsize=(8,5))
revenue_by_day.plot(kind="bar")
plt.title("Revenue by Day")
plt.savefig("revenue_by_day.png")
plt.show()
```

```python
# Tips by gender
plt.figure(figsize=(6,4))
sns.barplot(x="sex", y="tip", data=df)
plt.title("Average Tip by Gender")
plt.savefig("tips_by_gender.png")
plt.show()
```

```python
# Lunch vs dinner
plt.figure(figsize=(6,4))
sns.barplot(x="time", y="total_bill", data=df)
plt.title("Lunch vs Dinner Revenue")
plt.savefig("lunch_vs_dinner.png")
plt.show()
```

```python
# Smoker analysis
smoker_spending = df.groupby("smoker")["total_bill"].mean()
print(smoker_spending)
```

```python
# Group size analysis
group_spending = df.groupby("size")["total_bill"].mean()
print(group_spending)
```

```python
# Correlation
corr = df[["total_bill", "tip", "size"]].corr()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.savefig("correlation_heatmap.png")
plt.show()
```

```python
# Distribution
plt.figure(figsize=(8,5))
sns.histplot(df["total_bill"], kde=True)
plt.title("Bill Distribution")
plt.savefig("bill_distribution.png")
plt.show()
```
